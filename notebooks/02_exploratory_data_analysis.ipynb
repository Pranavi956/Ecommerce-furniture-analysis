```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset from previous step
df = pd.read_csv("../data/ecommerce_furniture_dataset_2024.csv")

# Clean price again (in case you run EDA standalone)
df['price'] = df['price'].replace(r'[$,]', '', regex=True).astype(float)
df = df.drop(columns=['originalPrice'], errors='ignore')
df['tagText'] = df['tagText'].fillna("Unknown")
```

```python
print(df.describe())
print(df.info())
```

```python
sns.histplot(df['price'], kde=True, bins=30)
plt.title("Distribution of Prices")
plt.show()

sns.histplot(df['sold'], kde=True, bins=30, color='green')
plt.title("Distribution of Units Sold")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='tagText')
plt.title("Tag Text Distribution")
plt.show()
```

```python
sns.scatterplot(x='price', y='sold', data=df)
plt.title("Price vs. Units Sold")
plt.show()
```

```python
corr = df[['price', 'sold']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
